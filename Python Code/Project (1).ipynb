```python
conditions_df = pd.read_excel("h241.xlsx")

# Preview first few rows
conditions_df.head()
```

```python
conditions_df.columns.tolist()
```

```python
# Filter rows where ICD10CDX starts with any chronic code
chronic_conditions = ('E11', 'I10', 'J45', 'I25', 'N18', 'J44')

# Drop NA just in case
conditions_df = conditions_df.dropna(subset=['ICD10CDX'])

# Filter only chronic patients
chronic_df = conditions_df[conditions_df['ICD10CDX'].str.startswith(chronic_conditions)]

# Get unique DUPERSIDs (patients) with chronic conditions
chronic_ids = chronic_df['DUPERSID'].unique()

# Filter main df for chronic patients only
df_chronic = df_filtered[df_filtered['DUPERSID'].isin(chronic_ids)]

# Show final output
df_chronic.head()
```

```python
# Load inpatient stays file
ip_df = pd.read_excel("h239d.xlsx")

# Preview first few rows
ip_df.head()
```

```python
ip_df.columns.tolist()
```

```python
# Keep only relevant columns
ip_visits = ip_df[["DUPERSID", "EVNTIDX", "IPBEGYR", "IPBEGMM"]].copy()

# Create a visit date column (for Recency)
ip_visits["Visit_Date"] = pd.to_datetime(
    ip_visits["IPBEGYR"].astype(str) + "-" + ip_visits["IPBEGMM"].astype(str).str.zfill(2) + "-01"
)

# Drop missing dates if any
ip_visits.dropna(subset=["Visit_Date"], inplace=True)

# Preview
ip_visits.head()
```

```python
# Load ER visit file
er_df = pd.read_excel("h239e.xlsx")

# Preview first few rows
er_df.head()
```

```python
er_df.columns.tolist()
```

```python
# Keep only relevant columns
er_visits = er_df[["DUPERSID", "EVNTIDX", "ERDATEYR", "ERDATEMM"]].copy()

# Create visit date column
er_visits["Visit_Date"] = pd.to_datetime(
    er_visits["ERDATEYR"].astype(str) + "-" + er_visits["ERDATEMM"].astype(str).str.zfill(2) + "-01"
)

# Drop any invalid dates
er_visits.dropna(subset=["Visit_Date"], inplace=True)

# Preview
er_visits.head()
```

```python
# Load office-based visit file
office_df = pd.read_excel("h239g.xlsx")

# Preview first few rows
office_df.head()
```

```python
office_df.columns.tolist()
```

```python
# Keep only relevant columns
office_visits = office_df[["DUPERSID", "EVNTIDX", "OBDATEYR", "OBDATEMM"]].copy()

# Create visit date column
office_visits["Visit_Date"] = pd.to_datetime(
    office_visits["OBDATEYR"].astype(str) + "-" + office_visits["OBDATEMM"].astype(str).str.zfill(2) + "-01"
)

# Drop invalid dates
office_visits.dropna(subset=["Visit_Date"], inplace=True)

# Preview
office_visits.head()
```

```python
# Load the Condition-Event Linkage file
clnk_df = pd.read_excel("h239if1.xlsx", engine="openpyxl")

# Preview the first few rows
clnk_df.head()
```

```python
clnk_df.columns.tolist()
```

```python

# Define chronic ICD-10 codes
chronic_conditions = ('E11', 'I10', 'J45', 'I25', 'N18', 'J44')

# Drop NA values in ICD codes
conditions_df = conditions_df.dropna(subset=['ICD10CDX'])

# Filter for chronic conditions
chronic_df = conditions_df[conditions_df['ICD10CDX'].str.startswith(chronic_conditions)]

# Get list of chronic CONDIDX
chronic_condidx_list = chronic_df['CONDIDX'].unique()

# Filter CLNK for only chronic-related visits
clnk_chronic = clnk_df[clnk_df['CONDIDX'].isin(chronic_condidx_list)]

# Preview
clnk_chronic.head()
```

```python
# Step 1: Filter each visit set using chronic EVNTIDXs
chronic_evnt_ids = clnk_chronic['EVNTIDX'].unique()

ip_visits_chronic = ip_visits[ip_visits['EVNTIDX'].isin(chronic_evnt_ids)]
er_visits_chronic = er_visits[er_visits['EVNTIDX'].isin(chronic_evnt_ids)]
office_visits_chronic = office_visits[office_visits['EVNTIDX'].isin(chronic_evnt_ids)]

# Step 2: Combine into one master visit log
all_visits_chronic = pd.concat([ip_visits_chronic, er_visits_chronic, office_visits_chronic])

# Step 3: Preview
all_visits_chronic.head()
```

```python
# Step 1: Group visits by DUPERSID
rfm_df = all_visits_chronic.groupby("DUPERSID").agg(
    Recency=("Visit_Date", lambda x: (pd.Timestamp("2022-12-31") - x.max()).days),
    Frequency=("Visit_Date", "count")
).reset_index()

# Step 2: Add Monetary from df_chronic
rfm_df = rfm_df.merge(df_chronic[["DUPERSID", "Total_Expenditure"]], on="DUPERSID", how="left")

# Preview
rfm_df.head()
```

```python
# Score RFM
rfm_df['R_Score'] = pd.qcut(rfm_df['Recency'], 3, labels=[3, 2, 1]).astype(int)
rfm_df['F_Score'] = pd.qcut(rfm_df['Frequency'].rank(method='first'), 3, labels=[1, 2, 3]).astype(int)
rfm_df['M_Score'] = pd.qcut(rfm_df['Total_Expenditure'].rank(method='first'), 3, labels=[1, 2, 3]).astype(int)

# Create RFM Segment
rfm_df['RFM_Segment'] = rfm_df['R_Score'].astype(str) + rfm_df['F_Score'].astype(str) + rfm_df['M_Score'].astype(str)

# Show output
rfm_df.head()
```

```python
# Step: Define RFM Segment Mapping
def map_segment(segment):
    if segment == '333':
        return 'Champion'
    elif segment[0] == '3':
        return 'Loyal'
    elif segment[0] == '1':
        return 'At Risk'
    elif segment == '111':
        return 'Lost'
    elif segment[1] == '3':
        return 'Frequent Buyer'
    else:
        return 'Others'

# Apply to your RFM dataframe
rfm_df['Segment_Label'] = rfm_df['RFM_Segment'].apply(map_segment)

# Preview it
rfm_df[['DUPERSID', 'RFM_Segment', 'Segment_Label']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count of patients in each segment
segment_counts = rfm_df['Segment_Label'].value_counts().reset_index()
segment_counts.columns = ['Segment_Label', 'Count']

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Segment_Label', data=segment_counts, palette='viridis')
plt.title('Patient Distribution by RFM Segment', fontsize=16)
plt.xlabel('Number of Patients')
plt.ylabel('RFM Segment')
plt.tight_layout()
plt.show()
```
